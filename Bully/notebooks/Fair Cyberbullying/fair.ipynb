```python
import tensorflow.keras as keras
from tensorflow.keras import layers
import tensorflow as tf

model = get_dense_model()
model.load_weights('models/wiki_constrained.h5')

# Read data
kaggle_data = pd.read_csv('/home/mb05005/mudit/Bully/bert/cyberbullying_tweets.csv')
kaggle_data.head()

# Step 4: Preprocess Your Data to Get BERT Embeddings
# Assuming you have a function `get_bert_embeddings` for generating embeddings
comments = kaggle_data['tweet_text'].tolist()

# Generate BERT embeddings for your data
your_embeddings = get_bert_embeddings(comments)  # Make sure this function matches your setup

# Step 5: Make Predictions
predictions = model.predict(your_embeddings, batch_size=16)

# Convert predictions to binary classes (assuming a threshold of 0.5)
predicted_classes = (predictions > 0.5).astype(int).flatten()
```
